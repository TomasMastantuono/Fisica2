```python
import numpy as np
import matplotlib.pyplot as plt

#Defino las constantes con valores que me pintan
f0 = 1
m = 10
gamma = 2
k = 3
w = np.arange(0, 5, .001)

#Defino las funciones A y B
def A(w):
    return(
        (f0 / m) * ((gamma*w) / (((k/m) - w**2)**2 + (gamma*w)**2))
    )

def B(w):
    return(
        (f0 / m) * ( ( (k / m) - w**2 ) / ( ((k / m) - w**2)**2 + (gamma*w)**2 ) )
    )

#Hago el plot
plt.figure()
plt.plot(w, A(w), linestyle='-', color ='red', marker='None', label='A')

plt.axhline(0, color='lightblue')
plt.axvline(0, color='lightblue')
plt.axvline((k/m)**1/2, color='purple', linestyle='--')

plt.plot(w, B(w), linestyle='-', color ='blue', marker='None', label='B')

plt.legend()
plt.show()
```
